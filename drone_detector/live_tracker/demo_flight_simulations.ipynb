# Drone Tracking Demo Simulations

Run these cells to show screen-recordable 3D demos for the project README. The scenes are simulated demo behaviors: fast follow, orbit tracking, and safe standoff following.

```python
from pathlib import Path
from IPython.display import FileLink, HTML, display
import importlib
import matplotlib.pyplot as plt

import demo_flight_simulations as dfs
dfs = importlib.reload(dfs)  # important if the notebook kernel imported an older version
SCENARIOS = dfs.SCENARIOS
DEFAULT_OUTPUT_DIR = dfs.DEFAULT_OUTPUT_DIR
create_animation = dfs.create_animation

DEFAULT_OUTPUT_DIR.mkdir(parents=True, exist_ok=True)
print(DEFAULT_OUTPUT_DIR)

def export_demo(name):
    scenario = SCENARIOS[name]
    mp4 = DEFAULT_OUTPUT_DIR / f"{name}_tracking_demo.mp4"
    gif = DEFAULT_OUTPUT_DIR / f"{name}_tracking_demo.gif"
    for output in (mp4, gif):
        fig, anim = create_animation(scenario, output)
        plt.close(fig)
    display(HTML(f"<h3>{scenario.title}</h3><p>{scenario.description}</p>"))
    display(FileLink(str(mp4), result_html_prefix="Download MP4: "))
    display(FileLink(str(gif), result_html_prefix="Download GIF: "))
    return mp4, gif
```

## Demo 1: Fast Follow Tracking

This shows the controlled drone following a moving target using smoothed camera/depth estimates.

```python
export_demo("follow")
```

## Demo 2: Orbit Tracking

This shows the controlled drone orbiting the target while keeping it in the simulated camera line of sight.

```python
export_demo("orbit")
```

## Demo 3: Safe Standoff Follow

This shows the controlled drone following the target while holding a minimum separation distance.

```python
export_demo("standoff")
```

Run this if you want pre-rendered files to link from the README or include in a demo folder.

```python
for name in SCENARIOS:
    export_demo(name)
```
